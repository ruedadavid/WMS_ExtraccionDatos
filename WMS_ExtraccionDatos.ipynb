# Extracción de Datos desde Servicio WMS

## Función Extracción desde WMS
### La función retorna datos desde un servicio WMS, se requiere utilizar la librería OwsLib, precargada en el OSGeoLive 

#### Dependencias:

1. Se requieren las librerías:

1.1. OWSLib
1.1.1. Esta librería requiere la librería lxml

#### Variables requeridas:
- url = Corresponde a la dirección URL del servicio WMS


```python
# Importando librería principal
from owslib.wms import WebMapService
# Función para extraer nombre de capa y cuadro de coordenadas desde un servicio WMS
def WMS_Capas(url):
    # Asignación de los datos del servicio
    wms = WebMapService(url)
    # Obtener las capas disponibles del servicio
    Capas_wms = wms.contents    
    listado_capas = list(Capas_wms)
    # Ciclo para extraer los datos de interés desde los respectivos nodos
    for capa in listado_capas:        
        print ('{};{}' .format(Capas_wms.get(capa).title, Capas_wms.get(capa).boundingBoxWGS84))
```

### Prueba con uno de los servicios

```python
# Declaración del sitio a visitar
paginaweb = 'https://neo.sci.gsfc.nasa.gov/wms/wms?'
# Llamado de función que consulta los datos de interés
WMS_Capas(paginaweb)
```

### Extracción de datos desde todos los servicios

```python
# Definición de diccionario con todos los servicios a consultar

urls = {'Metaspatial':'http://metaspatial.net/cgi-bin/ogc-wms.xml',
        'NASA':'https://neo.sci.gsfc.nasa.gov/wms/wms',
        'NOAA':'https://nowcoast.noaa.gov/arcgis/services/nowcoast/sat_meteo_imagery_time/MapServer/WMSServer',
        'EOX':'https://tiles.maps.eox.at/wms',
        'GEBCO':'https://www.gebco.net/data_and_products/gebco_web_services/web_map_service/mapserv'
       }
# Ciclo para visita a todos los servicios
for valor in urls:
    print('\n ### Servicio {}, {}:\n' .format(valor, urls[valor]))
    WMS_Capas(urls[valor])
```
